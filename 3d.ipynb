```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
print( cv2.__version__ )
```

```python
# Change paths to test images
imagepath1 = r"chariot1.jpg"
imagepath2 = r"chariot2.jpg"


def get_sift(img):
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # create SIFT object
    sift = cv2.SIFT_create()

    # detect SIFT features in both images
    keypoints, descriptors = sift.detectAndCompute(img1,None)

    sift_image = cv2.drawKeypoints(img, keypoints, img1, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    return keypoints, descriptors, sift_image


def get_orb(img):
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # create ORB object
    orb = cv2.ORB_create()

    # detect ORB features in both images
    keypoints, descriptors = orb.detectAndCompute(img1,None)

    orb_image = cv2.drawKeypoints(img, keypoints, img1, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    return keypoints, descriptors, orb_image

def get_matchcoords(matches, kp1, kp2):
    list_kp1 = []
    list_kp2 = []
    
    for mat in matches:
        img1_idx = mat.queryIdx
        img2_idx = mat.trainIdx

    # x - columns
    # y - rows
    # Get the coordinates
        (x1, y1) = kp1[img1_idx].pt
        (x2, y2) = kp2[img2_idx].pt

    # Append to each list
        list_kp1.append((x1, y1))
        list_kp2.append((x2, y2))
    
    return np.array(list_kp1), np.array(list_kp2)

def normalize(points):
    for row in points:
        row /= points[-1]
    return points

def make_homog(points):
    return np.vstack((points, np.ones((1,points.shape[1]))))
```

```python
img1 = cv2.imread(imagepath1)  
img2 = cv2.imread(imagepath2)
print(img1.shape, img2.shape)

# detect SIFT features in both images
keypoints_1, descriptors_1, sift_image1 = get_sift(img1)
keypoints_2, descriptors_2, sift_image2 = get_sift(img2)

plt.figure()
plt.subplot(1,2,1)
plt.imshow(sift_image1)

plt.subplot(1,2,2)
plt.imshow(sift_image2)
plt.show()


bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
# match descriptors of both images

matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)
matched_img = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:60], img2, flags=2)


match_coords1, match_coords2 = get_matchcoords(matches, keypoints_1, keypoints_2)
print(len(list1), len(list2))
print(len(matches))
plt.imshow(matched_img)
plt.show()
```
